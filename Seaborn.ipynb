```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Gráfico de linhas

```python
#index_col seta o index como a coluna data
#parse_date indica que cada dado no index é para ser compreendido como data

dados_fifa = pd.read_csv('fifa.csv', index_col= 'Date', parse_dates=True)

dados_fifa.head()
```

```python
#Plotando um gráfico de linha 
plt.figure(figsize=(16, 6))
plt.title('Fifa')
sns.lineplot(data=dados_fifa)
```

```python
#Selecionando dados de colunas específicas
plt.figure(figsize=(10, 6)) #definindo tamanho
plt.title('Brasil Vs Argentina') #definindo titulo
plt.xlabel('Data') #definindo titulo eixo x

#Definindo colunas específicas
sns.lineplot(data=dados_fifa['BRA'], label='Brasil')
sns.lineplot(data=dados_fifa['ARG'], label = 'Argentina')
```

# Gráficos de barras e mapa de calor

```python
dados_voos = pd.read_csv('flight_delays.csv')
dados_voos.head()
```

### Gráfico de barras

```python
#Plotando um gráfico de barras para a coluna AA
plt.figure(figsize = (10, 5))
plt.title('Delay de chegada por mês')

sns.barplot(x = dados_voos['Month'], y = dados_voos['AA'])
plt.ylabel('Delay em minutos') #Definindo o titulo do eixo y
plt.xlabel('Meses') #Definindo o titulo do eixo x
```

### Mapa de calor

```python
#Plotando um mapa de calor sem os valores (annot = False por padrão)
plt.figure(figsize = (10,6))
plt.title('Média de delay de chegada de cada companhia aérea')

sns.heatmap(data = dados_voos.iloc[:,1:])
plt.ylabel('Meses') #Nomeando o eixo y
plt.xlabel('CIA Aérea') #Nomeando o eixo x
```

```python
#Plotando um mapa de calor sem os valores (annot = True)
plt.figure(figsize = (10,6))
plt.title('Média de delay de chegada de cada companhia aérea')

sns.heatmap(data = dados_voos.iloc[:,1:], annot = True)
plt.ylabel('Meses') #Nomeando o eixo y
plt.xlabel('CIA Aérea') #Nomeando o eixo x
```

# Gráfico de dispersão

```python
seguro = pd.read_csv('insurance.csv')
seguro.head()
```

### Gráfico de dispersão

#### Sem linha de regressão

```python
#Criando um gráfico de dispersão
plt.figure(figsize = (10, 6))
plt.title('Valor de seguro por IMC')
plt.xlabel('IMC')
plt.ylabel('Valor de seguro')

sns.scatterplot(x = seguro['bmi'], y = seguro['charges'])
```

#### Com linha de regressão

```python
#Criando um gráfico de dispersão com linha de regressão
plt.figure(figsize = (10, 6))
plt.title('Valor de seguro por IMC')
plt.xlabel('IMC')
plt.ylabel('Valor de seguro')

sns.regplot(x = seguro['bmi'], y = seguro['charges'])
```

#### Adicionando mais uma variável à correlação

```python
#Identificando correlação entre 3 variáveis
plt.figure(figsize = (10, 6))
plt.title('Valor de seguro por IMC')
plt.xlabel('IMC')
plt.ylabel('Valor de seguro')

sns.scatterplot(x = seguro['bmi'], y = seguro['charges'], hue = seguro['smoker']) #adição do parâmetro hue
```

```python
list(seguro.columns)
```

#### Adicionando linhas de regreção para cada variável adicional

```python
#Adicionando mais linhas de regressão para cada condição
sns.lmplot(x = 'bmi', y = 'charges', hue = 'smoker', data = seguro)
```

#### Fixando uma das variáveis a um eixo principal

```python
#Fixando o eixo de uma das variáveis
plt.figure(figsize = (10, 6))
plt.title('Valor de seguro por IMC')
plt.xlabel('Fumante')
plt.ylabel('Valor de seguro')

sns.swarmplot(x = seguro['smoker'], 
              y = seguro['charges'],
             size = 2)
```
